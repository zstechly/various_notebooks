```python
import numpy as np
from scipy import signal, fftpack
from matplotlib import pyplot as plt
```

```python
# create the signal of interest.  What we want to do is take our signal
Fs = 12
x = np.zeros(16)
x[0:8] = 1
plt.plot(x)
plt.grid(True)
plt.title('Input Signal')
```

```python
sig_total_fft = fftpack.fft(x)
```

```python
plt.plot(((fftpack.fftshift(abs(sig_total_fft)))))
plt.grid(True)
plt.title('Frequency Components')
```

```python
sig_reconstruct = np.zeros(16)
t = np.arange(16)
for n in range(16):
    sig_reconstruct = sig_reconstruct + (sig_total_fft[n] * (np.cos(2*np.pi*n/16*t) + 1j*np.sin(2*np.pi*n/16*t)))
```

```python
sig_reconstruct
```

```python
plt.plot(1/16*sig_reconstruct.real)
plt.grid(True)
plt.title('Reconstructed signal scaled')
```

```python
fig, axes = plt.subplots(1,2)
sig_reconstruct = np.zeros(16)
t = np.arange(16)
for n in range(16):
    sig_reconstruct = sig_reconstruct + (sig_total_fft[n] * (np.cos(2*np.pi*n/16*t) + 1j*np.sin(2*np.pi*n/16*t)))
    axes[0].plot(sig_reconstruct.real)
    axes[1].plot(sig_reconstruct.imag)
axes[0].grid(True)
axes[0].title.set_text('Real component \n reconstructed')
axes[1].grid(True)
axes[1].title.set_text('Image components \n reconstructed')
```

```python
# individual tones
fig, axes = plt.subplots(1,2)
sig_reconstruct = np.zeros(16)
t = np.arange(16)
for n in range(16):
    sig_reconstruct = sig_reconstruct + (sig_total_fft[n] * (np.cos(2*np.pi*n/16*t) + 1j*np.sin(2*np.pi*n/16*t)))
    axes[0].plot((sig_total_fft[n] * (np.cos(2*np.pi*n/16*t) + 1j*np.sin(2*np.pi*n/16*t)).real))
    axes[1].plot((sig_total_fft[n] * (np.cos(2*np.pi*n/16*t) + 1j*np.sin(2*np.pi*n/16*t)).imag))
axes[0].grid(True)
axes[1].grid(True)
axes[0].title.set_text('Real Components')
axes[1].title.set_text('Imag Components')
```
