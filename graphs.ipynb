```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(10))
ax.set_title('Hello')
ax.set_xlabel('A number')
ax.set_ylabel('Another number')
ax.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(10),linewidth=5.0, alpha=0.25)
ax.set_title('Hello')
ax.set_xlabel('A number')
ax.set_ylabel('Another number')
ax.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(10),linewidth=5.0, alpha=0.25)
ax.set_title('Hello')
ax.set_xlabel('A number')
ax.set_ylabel('Another number')
ax.grid(True)
x = 5
ax.text(2,4,f'number {x}')
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(np.arange(10),linewidth=5.0, alpha=0.25)
ax[0].set_title('Hello')
ax[0].set_xlabel('A number')
ax[0].set_ylabel('Another number')
ax[0].grid(True)
x = 5
ax[0].text(2,4,f'number {x}')
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(2)
ax[0].plot(np.arange(10),linewidth=5.0, alpha=0.25)
ax[0].set_title('Hello')
ax[0].set_xlabel('A number')
ax[0].set_ylabel('Another number')
ax[0].grid(True)
x = 5
ax[0].text(2,4,f'number {x}')
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()
ax.plot(100*np.random.rand(20))

# Use automatic StrMethodFormatter
ax.yaxis.set_major_formatter('${x:1.2f}')

ax.yaxis.set_tick_params(which='major', labelcolor='green',
                         labelleft=False, labelright=True)

```
