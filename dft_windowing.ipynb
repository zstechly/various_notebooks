The point of windowing is to drive the beginning and end of a sampled sequence to a common point, to avoid leakage as shown in the dft leakage notebook.  A non perfect number of cycles of a signal result in leakage.  Also weirdly of note, Hamming is Hanning on a pedastal 

```python
from scipy import signal
from scipy.fftpack import fft, fftshift
import matplotlib.pyplot as plt
```

```python
# make some windows
window = signal.hamming(51)
plt.plot(window)
plt.title("Hamming window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
plt.grid(True)
```

```python
window = signal.hanning(51)
plt.plot(window)
plt.title("Hanning window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
plt.grid(True)
```

Hamming, Hanning, Triangular, Chebychev, kaiser.  All tradeoffs between energy in main lobe, main lobe width, side lobe suppression.  This is just filtering!
