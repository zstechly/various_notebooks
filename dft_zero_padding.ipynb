Explore zero padding

```python
import numpy as np
from scipy import signal, fftpack
from matplotlib import pyplot as plt
```

```python
n = 16
fs = 1
fundamental_freq = fs / (n-1)
f = fundamental_freq*3
t = np.arange(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t)
sig_in_fft =fftpack.fft(sig_in)
```

```python
plt.plot(sig_in_fft.real)
plt.grid(True)
```

Now zero pad

```python
N=64
zero_vector = np.zeros(N)
sig_in_zero = np.append(sig_in, zero_vector)
sig_zero_fft = fftpack.fft(sig_in_zero)
```

```python
plt.plot(sig_zero_fft.real)
plt.grid(True)
```

This gives a better representation of the CFT but not give more resolution (aka two signals close to one another will not be clearer)

```python
n = 16
fs = 1
fundamental_freq = fs / (n-1)
f = fundamental_freq*3
f2 = fundamental_freq*3.1
t = np.arange(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t) + np.cos(2*np.pi*f2/fs*t) + 1j*np.sin(2*np.pi*f2/fs*t)
sig_in_fft =fftpack.fft(sig_in)
```

```python
plt.plot(sig_in_fft.real)
plt.grid(True)
```

```python
N=64
zero_vector = np.zeros(N)
sig_in_zero = np.append(sig_in, zero_vector)
sig_zero_fft = fftpack.fft(sig_in_zero)
plt.plot(sig_zero_fft.real)
plt.grid(True)
```

Now let's increase signal size and see if we can see those two signals

```python
n = 1024
fs = 1
fundamental_freq = fs / (15)
f = fundamental_freq*3
f2 = fundamental_freq*3.1
t = np.arange(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t) + np.cos(2*np.pi*f2/fs*t) + 1j*np.sin(2*np.pi*f2/fs*t)
sig_in_fft =fftpack.fft(sig_in)
plt.plot(sig_in_fft.real[100:300])
plt.grid(True)
```
