```python
import numpy as np
from scipy import signal, fftpack
from matplotlib import pyplot as plt
```

```python
n = 1024
fs = 1000
fundamental_freq = fs / (n-1)
f = 300
t = np.arange(n)
noise = np.random.randn(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t)+noise
sig_fft = 20*np.log10(fftpack.fft(sig_in))
sig_fft = sig_fft - np.max(sig_fft)
```

```python
plt.plot(sig_fft)
plt.grid(True)
```

Every doubling of N results in 3dB increase in SNR
