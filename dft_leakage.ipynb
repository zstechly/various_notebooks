```python
import numpy as np
from scipy import signal, fftpack
from matplotlib import pyplot as plt
```

Create a perfect tone to put all the energy in one bin

```python
n = 16
fs = 1
fundamental_freq = fs / (n-1)
f = fundamental_freq*2
t = np.arange(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t)
sig_in_fft =fftpack.fft(sig_in)
```

```python
plt.plot(sig_in.real)
plt.grid(True)
```

```python
plt.plot(20*np.log10(abs(sig_in_fft)))
plt.grid(True)
```

Now create a less than perfectly repeating tone

```python
n = 16
fs = 1
fundamental_freq = fs / (n)
f = fundamental_freq*2
t = np.arange(n)
sig_in = np.cos(2*np.pi*f/fs*t) + 1j*np.sin(2*np.pi*f/fs*t)
sig_in_fft =fftpack.fft(sig_in)
plt.plot(sig_in.real)
plt.grid(True)
```

```python
plt.plot(20*np.log10(abs(sig_in_fft)))
plt.grid(True)
```
