```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rxa_temp = pd.read_csv('./Desktop/nominal_temp_results/read_temps_data_rxa.csv', delimiter=',')
tx_temp = pd.read_csv('./Desktop/nominal_temp_results/read_temps_data_tx.csv', delimiter=',')
rxb_temp = pd.read_csv('./Desktop/nominal_temp_results/read_temps_data_rxb.csv', delimiter=',')
```

```python
rxa_temp = rxa_temp.drop(['chan2','chan4', 'chan5','chan6','chan7'], axis=1)
tx_temp = tx_temp.drop(['chan2','chan4', 'chan5','chan6','chan7'], axis=1)
rxb_temp = rxb_temp.drop(['chan2','chan4', 'chan5','chan6','chan7'], axis=1)
```

```python
plt.figure()
rxa_temp.plot(x='time',y=['chan0','chan1','chan3'])
plt.grid(True)
plt.ylabel('DegC')
plt.xlabel('Time')
plt.ylim([0,45])
plt.title('RXA FPGA SDM Temperature Measurements')
plt.savefig('./Desktop/nominal_temp_results/rxa_graph.png',dpi=200)
```

```python
plt.figure()
tx_temp.plot(x='time',y=['chan0','chan1','chan3'])
plt.grid(True)
plt.ylabel('DegC')
plt.xlabel('Time')
plt.ylim([0,45])
plt.title('TX FPGA SDM Temperature Measurements')
plt.savefig('./Desktop/nominal_temp_results/tx_graph.png',dpi=200)
```

```python
plt.figure()
rxb_temp.plot(x='time',y=['chan0','chan1','chan3'])
plt.grid(True)
plt.ylabel('DegC')
plt.xlabel('Time')
plt.ylim([0,45])
plt.title('RXB FPGA SDM Temperature Measurements')
plt.savefig('./Desktop/nominal_temp_results/rxb_graph.png',dpi=200)
```
