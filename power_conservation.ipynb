Using parsavel's theorum to conserve energy between frequency and time domains

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import signal, fftpack
```

```python
# create a signal 
Fs = 100;
F = 5;
t = np.arange(4096)
scale = 100
sig_in = scale* (np.cos(2*np.pi*F/Fs*t) + 1j*np.sin(2*np.pi*F/Fs*t))
```

```python
sig_time_pwr = sig_in.real**2 + sig_in.imag**2
```

```python
plt.plot(sig_in[0:100].real)
plt.plot(sig_in[0:100].imag)
```

We have i^2+q^2, lets calculate sum(i^2+q^2) in the time domain

```python
sig_time_pwr.sum()
```

Convert time domain signal to frequency domain

```python
sig_fft = fftpack.fftshift(fftpack.fft(sig_in))
```

```python
freq_axis = np.linspace(-50,50,sig_fft.size)
plt.plot(freq_axis,sig_fft.real)
plt.grid(True)
```

Per https://en.wikipedia.org/wiki/Parseval%27s_theorem we should see that the total power in the time domain = 1/N * sum(i^2+q^2) in frequency domain, which we see here

```python
sig_fft_pwr = (sig_fft.real**2 + sig_fft.imag**2)
sig_fft_pwr.sum() / sig_fft.size
```
