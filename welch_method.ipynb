```python
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
```

lam data is 524288.  Fs is 5.125e9.  wintype = rectangular with length 1024.  Noverlap
Noverlap=512, twosided is true.
you probably need detrend=False

```python
# create a cos/ sin wave
Fs = 5.125e9
f = 200e6
t = np.arange(524288)
sig_in = np.cos(2*np.pi*f/Fs*t) + 1j*np.sin(2*np.pi*f/Fs*t)+1*np.random.randn(524288)
faxis,pxx = signal.welch(sig_in,fs=Fs,window=np.ones(1024), nfft=1024, noverlap=512,detrend=False, return_onesided=False)
```

```python
plt.plot(faxis, 10*np.log10(pxx))
plt.grid(True)
plt.xlabel('Frequency (Hz)')
plt.ylabel('PXX')
```
